```python
import mysql.connector

# Conexión a la base de datos
cnx = mysql.connector.connect(
    user="root",
    password="AlumnaAdalab",
    host="127.0.0.1",
    database="CineParadiso",
    auth_plugin='mysql_native_password'
)

mycursor = cnx.cursor()

# Datos de ejemplo para insertar
data_film_cast_oscars = [
    (1, 1, 1),  # (film_id, cast_id, oscar_id)
    (2, 2, 2),
    (3, 3, 3)
]

# Consulta para insertar datos
query = 'INSERT INTO film_cast_oscars (film_id, cast_id, oscar_id) VALUES (%s, %s, %s)'

# Función para verificar la existencia de claves foráneas antes de insertar
def check_foreign_keys(cursor, film_id, cast_id, oscar_id):
    cursor.execute("SELECT film_id FROM film WHERE film_id = %s", (film_id,))
    if cursor.fetchone() is None:
        return False

    cursor.execute("SELECT cast_id FROM `cast` WHERE cast_id = %s", (cast_id,))
    if cursor.fetchone() is None:
        return False

    cursor.execute("SELECT oscar_id FROM oscars WHERE oscar_id = %s", (oscar_id,))
    if cursor.fetchone() is None:
        return False

    return True

try:
    for record in data_film_cast_oscars:
        if check_foreign_keys(mycursor, record[0], record[1], record[2]):
            mycursor.execute(query, record)
        else:
            print(f"Invalid foreign key combination: {record}")
    cnx.commit()
    print("Datos insertados correctamente en la tabla film_cast_oscars.")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    print("Error Code:", err.errno)
    print("SQLSTATE", err.sqlstate)
    print("Message", err.msg)
finally:
    mycursor.close()
    cnx.close()
```

```python
import mysql.connector

# Conexión a la base de datos
cnx = mysql.connector.connect(
    user="root",
    password="AlumnaAdalab",
    host="127.0.0.1",
    database="CineParadiso",
    auth_plugin='mysql_native_password'
)

mycursor = cnx.cursor()

# Verificar existencia de las tablas y crearlas si no existen

create_film_table = """
CREATE TABLE IF NOT EXISTS film (
    film_id INT AUTO_INCREMENT PRIMARY KEY,
    idOwner VARCHAR(30),
    title VARCHAR(100),
    type VARCHAR(30),
    year VARCHAR(10),
    genre VARCHAR(30),
    screenwriter VARCHAR(50),
    management VARCHAR(50),
    plot VARCHAR(1000),
    length INT,
    rating VARCHAR(10)
);
"""

create_cast_table = """
CREATE TABLE IF NOT EXISTS `cast` (
    cast_id INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(50),
    birth VARCHAR(30),
    award VARCHAR(500),
    nomination VARCHAR(500),
    known_for VARCHAR(500)
);
"""

create_oscars_table = """
CREATE TABLE IF NOT EXISTS oscars (
    oscar_id INT AUTO_INCREMENT PRIMARY KEY,
    date VARCHAR(30),
    best_film VARCHAR(50),
    best_management VARCHAR(40),
    best_actor VARCHAR(40),
    best_actress VARCHAR(40)
);
"""

create_film_cast_oscars_table = """
CREATE TABLE IF NOT EXISTS film_cast_oscars (
    id_film_cast_oscars INT AUTO_INCREMENT PRIMARY KEY,
    film_id INT,
    cast_id INT,
    oscar_id INT,
    FOREIGN KEY (film_id) REFERENCES film(film_id)
    ON DELETE CASCADE ON UPDATE CASCADE,
    FOREIGN KEY (cast_id) REFERENCES `cast`(cast_id)
    ON DELETE CASCADE ON UPDATE CASCADE,
    FOREIGN KEY (oscar_id) REFERENCES oscars(oscar_id)
    ON DELETE CASCADE ON UPDATE CASCADE
);
"""

try:
    mycursor.execute(create_film_table)
    print("Tabla 'film' verificada/creada.")
    mycursor.execute(create_cast_table)
    print("Tabla 'cast' verificada/creada.")
    mycursor.execute(create_oscars_table)
    print("Tabla 'oscars' verificada/creada.")
    mycursor.execute(create_film_cast_oscars_table)
    print("Tabla 'film_cast_oscars' verificada/creada.")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    print("Error Code:", err.errno)
    print("SQLSTATE", err.sqlstate)
    print("Message:", err.msg)
finally:
    mycursor.close()
    cnx.close()
```
