```python
# Tabla FILM_CAST_OSCARS

cnx = mysql.connector.connect(
    user="root",
    password="AlumnaAdalab",
    host="127.0.0.1",
    database="CineParadiso",
    auth_plugin='mysql_native_password'
)

mycursor = cnx.cursor()


query = ("""
         CREATE TABLE film_cast(
    id_film_cast INT AUTO_INCREMENT PRIMARY KEY,
    cast_id INT,
    idOwner VARCHAR (10),
    name VARCHAR (50),
    film_id INT, 
    title VARCHAR(50),
    FOREIGN KEY (film_id) REFERENCES film (film_id)
    ON DELETE CASCADE ON UPDATE CASCADE,
    FOREIGN KEY (cast_id) REFERENCES cast (cast_id)
    ON DELETE CASCADE ON UPDATE CASCADE);
    """)


try:
    mycursor.execute(query)

    print(mycursor)
    
except mysql.connector.Error as err:
    print(err)
    print("Error Code:", err.errno)
    print("SQLSTATE", err.sqlstate)
    print("Message", err.msg)
    
cnx.close()
```

```python

cnx = mysql.connector.connect(
    user="root",
    password="AlumnaAdalab",
    host="127.0.0.1",
    database="CineParadiso",
    auth_plugin='mysql_native_password'
)

mycursor = cnx.cursor()

## Leer el CSV donde tengo la info:

data_movies = pd.read_csv('../info_tablas_csv/Oscars.csv')

# Reemplazar NaN con None para asegurar que los valores nulos se insertan correctamente
data_movies = data_movies.where(pd.notnull(data_movies), None)


query = 'INSERT INTO Oscars (date, best_film, best_management, best_actor, best_actress) VALUES (%s,%s,%s,%s,%s)'


for _, row in data_movies.iterrows():
    
    try:
        mycursor.execute(query, tuple(row))
    
    except mysql.connector.Error as err:
        print(f"Error insertando fila {row['id']}: {err}")
        
## Confirmar los cambios en la base de datos

cnx.commit()

## Cerrar el cursor y la conexión

mycursor.close()
cnx.close()


```

```python
import mysql.connector
import pandas as pd

# Conexión a la base de datos
cnx = mysql.connector.connect(
    user="root",
    password="AlumnaAdalab",
    host="127.0.0.1",
    database="CineParadiso",
    auth_plugin='mysql_native_password'
)

mycursor = cnx.cursor()

# Leer el CSV donde tienes la información
data_movies = pd.read_csv('../info_tablas_csv/tabla_actor_film_conid.csv')

# Reemplazar NaN con None para asegurar que los valores nulos se insertan correctamente
data_movies = data_movies.where(pd.notnull(data_movies), None)

# Verificar que el reemplazo se realizó correctamente
print(data_movies.head())

# Definir la consulta de inserción
query = ('INSERT INTO film_cast (cast_id,idOwner,name,film_id,title) VALUES (%s,%s,%s,%s,%s)')

# Convertir DataFrame a una lista de tuplas 
data_tuples = [tuple(x) for x in data_movies.to_numpy()]

# Iterar sobre cada tupla e insertar en la base de datos
for row in data_tuples:
    try:
        mycursor.execute(query, row)
    except mysql.connector.Error as err:
        print(f"Error insertando fila {row[0]}: {err}")

# Confirmar los cambios en la base de datos
cnx.commit()

# Cerrar el cursor y la conexión
mycursor.close()
cnx.close()
```
