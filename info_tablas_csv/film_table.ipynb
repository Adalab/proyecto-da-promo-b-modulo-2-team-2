```python
import pandas as pd
```

```python
action_elenco = pd.read_csv('action_elenco.csv')
comedy_elenco = pd.read_csv('Comedy_elenco.csv')
drama_elenco = pd.read_csv('Drama_elenco.csv')
fantasy_elenco = pd.read_csv('Fantasy_elenco.csv')
horror_elenco = pd.read_csv('Horror_elenco.csv')
mystery_elenco = pd.read_csv('Mystery_elenco.csv')
romance_elenco = pd.read_csv('Romance_elenco.csv')
thriller_elenco = pd.read_csv('thriller_elenco.csv')
```

```python
# lista de archivos csv 
csv_genre = ['action_elenco.csv','Comedy_elenco.csv','Drama_elenco.csv','Fantasy_elenco.csv','Horror_elenco.csv','Mystery_elenco.csv','Romance_elenco.csv','thriller_elenco.csv']
```

```python
# Leer cada uno de los archivos csv en un DataFrame y almacenarlos en una lista 
dfs = [pd.read_csv(file) for file in csv_genre]

# Concatenar todos los DataFrames en uno solo 
df_final = pd.concat (dfs, axis=0, ignore_index = True)
```
