```python
# Importar librerías para web scraping y manipulación de datos
# -----------------------------------------------------------------------
from bs4 import BeautifulSoup
import requests

# Importar librerías para manipulación y análisis de datos
# -----------------------------------------------------------------------
import pandas as pd

# Importar librerías para procesamiento de texto
# -----------------------------------------------------------------------
import re
```

```python
url = 'https://es.wikipedia.org/wiki/Premios_%C3%93scar'

llamada = requests.get(url)

llamada.status_code

```

```python
sopa = BeautifulSoup(llamada.content, 'html.parser')

tabla = sopa.find_all('table')

# Para saber si hay más de una tabla en Wikipedia:

len(tabla)
```

```python
# Para saber lo que hay en cada tabla:

print("En la tabla 1 tenemos: \n ", tabla[0])
print("\n---------------------\n")

print("En la tabla 2 tenemos: \n", tabla[1])
print("\n---------------------\n")

print("En la tabla 3 tenemos: \n", tabla[2])
print("\n---------------------\n")

print("En la tabla 4 tenemos: \n", tabla[3])



```

```python
# Nuestra tabla, es la tabla 3

tabla_premios = tabla[2]

tabla_premios
```

```python
lista_encabezados = tabla_premios.find_all('th')

lista_encabezados
```

```python
# Para sacar los encabezados:

encabezados = []

for columna in lista_encabezados:
    encabezados.append(columna.text.replace('\n', ' ').strip())
    

encabezados
```

```python
# Para sacar las listas:

filas = tabla_premios.find_all('tr')

filas
```

```python
resultados = []


for fila in filas[52:68]:
    
    separacion = []
    fila_limpia = fila.text.split('\n')
    
    for element in fila_limpia:
        if element != '':
            
            if element != '—':
                separacion.append(element.replace('\n', '').strip())

        
        
    resultados.append(separacion)

resultados
```

```python
df = pd.DataFrame(resultados)

df

df_limpio = df.drop(columns=[0, 6, 7, 8, 9])

df_limpio
```

```python
encabezados = ['Fecha de ceremonia', 'Mejor película', 'Mejor dirección', 'Mejor actor', 'Mejor actriz']

df_limpio.columns = encabezados

df_limpio

```

```python
df_limpio.to_csv("../info_tablas_csv/Oscars.csv")
```
